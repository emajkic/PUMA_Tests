```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

print("Sucessful Imports")
```

```python
# Set data directory and voltage values
ddir = r'Rscan_VA'
VG = np.arange(0, 10, 2)  
Na = len(VG)  
Tim = np.arange(1, 5001) * 2e-3  # Time, window for pulse in microseconds

WC = np.zeros((5000, Na))  # Cathode pulses
WA = np.zeros((5000, Na))  # Anode pulses
QM = np.zeros((Na, 2))  # Anode and cathode charges

i = 0

for nr in range(Na):
    # Read data
    fileName = f"Rscan_VA0_VG{i}.txt"  # File to open
    i = i+2
    TD = pd.read_table(fileName, header=None)  # Read file as table
    dt = TD.iloc[0, 0]  # Time sample in ns
    DC = TD.iloc[1:, :10].values  # Cathode waveforms
    DA = TD.iloc[1:, 10:20].values  # Anode waveforms

    # Average pulses
    DCa = np.mean(DC, axis=1)  # Average over 10 cathode waveforms
    DAa = np.mean(DA, axis=1)  # Average over 10 anode waveforms
    DCa -= np.mean(DCa[:4000])  # Subtract baselines
    DAa -= np.mean(DAa[:4000])  
    DCa = DCa[4500:9500]  # Take 5000 significant samples
    DAa = DAa[4500:9500]
    
    WC[:, nr] = DCa  # Store pulse
    WA[:, nr] = DAa
    
    # Charge calculation 
    QM[nr, 0] = 1e9 * np.sum(DCa) * dt / 220  # Cathode charge, pC
    QM[nr, 1] = -1e9 * np.sum(DAa) * dt / 220  # Anode inversed charge, pC
```

```python
# Output plots
plt.figure('Measured signals')
plt.plot(Tim, np.hstack([WC, WA]))
plt.xlabel('Time, us')
plt.ylabel('Signal, V')
plt.grid(True)

plt.figure('Input charge')
plt.plot(VG, QM, '-o')
plt.xlabel('Grid bias, V')
plt.ylabel('Charge, pC')
plt.grid(True)

plt.show()
```
